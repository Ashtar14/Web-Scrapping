```python
from bs4 import BeautifulSoup
import lxml
import html5lib
import requests
import pandas as pd
results=pd.DataFrame()
l=["https://arifpucit.github.io/bss2/CA.html","https://arifpucit.github.io/bss2/index.html","https://arifpucit.github.io/bss2/SP.html"]
z=0
for i in l:

    resp=requests.get(i)

    soup=BeautifulSoup(resp.text,"lxml")

    price=soup.find_all("p",class_="price green")

    p=[]
    for pr in price:
        p.append(pr.text)
    
    booknames=soup.find_all('p',class_='book_name')

    names=[]
    for i in booknames:
        names.append(i.text)

    bnames=[]
    auth=[]
    for i in names:
        a=i.split(sep="By")
        if len(a)>=2:
            bnames.append(a[0])
            auth.append(a[1])
        else:
            bnames.append(a[0])
            auth.append("nothing")
        
    import pandas as pd

    df1=pd.DataFrame(bnames,auth)

    df1.columns=["Books"]

    df1["Authors"]=df1.index 

    df1.reset_index(drop=True,inplace=True)

    PKR=[]
    for i in p:
        a=i.split(sep="Rs.")
        PKR.append(a[1])

    df1["Price"]=PKR

    a=soup.find_all("div",class_="align-left")
    s=[]
    for i in a:
        c=i.text.split(sep="\n")
        s.append(c)
    

    Stock=[] 
    for c in s:
        Stock.append(c[2])

    review=[] 
    for c in s:
        review.append(c[4])

    df1["Stock Status"]=Stock
    df1["Reviews"]=review

    a=soup.find_all("div",class_="align-left")
    rating=[]
    for i in a:
        c=len(i.find_all("span","fa fa-star"))
        rating.append(c)
    df1["Rattings"]=rating
    r=[]
    for i in df1["Reviews"]:
        c=i.split(sep=" ")
        r.append(c[0])
    df1["Reviews"]=r

    a=soup.find_all('p',class_='book_name')
    l=[]
    for i in a:
        c=i.find_all('a')
        for e in c:
            l.append(e)
    e=[]
    for i in l:
        e.append(i.attrs)
    URL=[]
    for i in e:
        URL.append(i["href"])
    df1["URL"]=URL
    if z ==  0:
        results=df1
    else:
        results=results.append(df1,ignore_index=True)
    z=z+1
    
results
```
